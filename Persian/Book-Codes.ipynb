```python
class Person:
    def __init__(self, name):
        self.__name = name
p1 = Person('ali')
print(p1.name)
# output:
    # AttributeError
```

```python
class Person:
    def __init__(self, name):
        self.__name = name

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name
```

```python
class Person:
    def say_hello(self):
        print("Hello")
p = Person()
p.say_hello()
# output:
    # Hello
```

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")
p1 = Dog()
p1.speak()
p1.bark()
# output:
    # Animal speaks
    # Dog barks
```

```python
class Person:
    def __init__(self, name):
        self.name = name
    def __str__(self):
        return f"Name: {self.name}"
p = Person('Texas')
print(p)
# output:
    # Name: Texas
```
